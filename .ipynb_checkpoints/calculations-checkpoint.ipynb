# Observable: Flux

```python
def planckSpectralRadianceVsWavelength(lam, temp):
    """
    Calculates spectral radiance (B_lambda) of a blackbody.

    Args:
        lam (Quantity): Wavelength(s) with units (e.g., 500*u.nm).
        temp (Quantity): Temperature with units (e.g., 5800*u.K).

    Returns:
        float or array: Spectral radiance in SI units (W / (m^2 sr m)).
    """
    l = lam.si.value
    t = temp.si.value
    h = const.h.si.value
    k = const.k_B.si.value
    c = const.c.si.value
    coeff = 2.0 * h * c**2/l**5
    exponent = h * c / (k * t * l)
    out = coeff / np.expm1(exponent)
    return out
```

```python
def radiantFluxBandpass(temp, lamMin, lamMax):
    """
    Calculates the radiant flux by integrating Planck's Law.
    """
    wavelengths = np.logspace(np.log10(lamMin.si.value), np.log10(lamMax.si.value), 500) * u.m
    integrand = planckSpectralRadianceVsWavelength(wavelengths, temp)
    out = np.pi * integrate.trapezoid(integrand, x=wavelengths.si.value)
    return out
```

```python
def luminosityBandpass(temp, radius, lamMin, lamMax):
    """
    Calculates the luminosity from the radiant flux and radius.
    """
    M = radiantFluxBandpass(temp, lamMin, lamMax)
    surfaceArea = 4 * np.pi * radius**2
    surfaceArea = surfaceArea.si.value
    L = M * surfaceArea
    return L
```

```python
def luminositySB(temp, radius):
    """
    Calculates the luminosity of a blackbody using the Stefan Boltzmann law
    """
    sigma = const.sigma_sb.si.value
    out = 4 * np.pi * radius.si.value**2 * sigma * temp.si.value**4
    return out
```

```python
def blackBodyTemp(incidentFlux, emissivity = 1.0):
    sigma = const.sigma_sb.si.value
    return (incidentFlux/(emissivity*sigma))**0.25
```

Currie 2023 gives the following definition for contrast:
$$
C_{\mathrm{IR}, \lambda} \sim \frac{F_{\lambda, p}(T, X) r_{\mathrm{p}}^2}{F_{\lambda, s}(T) r_{\mathrm{s}}^2}
$$

## Planet Luminosity and Contrast as a Function of Stellar Luminosity

```python
def scalarLuminosityToMass(lstar):
    """
    Calculates the stellar from the luminosity based on Halm's relation (see the Wikipedia article https://en.wikipedia.org/wiki/Mass-luminosity_relation and references therein).

    Args:
        lstar (number): Luminosity in units of $L_\\odot$.

    Returns:
        Stellar mass (number): Mass in units of $M_\\odot$.
    """
    if lstar < 0.0:
        return 0.0
    elif lstar < 0.03418801:
        return (lstar/0.23817511001034142)**(1/2.3)
    elif lstar < 16.0:
        return (lstar)**(1/4)
    elif lstar < 1744956:
        return (lstar/np.sqrt(2))**(1/3.5)
    else:
        return lstar/3.2e5
        
luminosityToMass = np.vectorize(scalarLuminosityToMass)
```

```python
def distanceToHostStar(period, mstar):
    """
    Calculate radius of a circular orbit around a star of mass `mstar` and duration of `period`.

    Args:
        period (Quantity): period of the orbit in days
        mstar (Quantity): stellar mass

    Returns:
        Radius of the orbit in au
    """

    out = (const.G * mstar * period**2/(4*np.pi**2))
    out = out**(1/3)
    return out.to(u.AU)
    
```

```python
# check that we get close to 1 au for our own solar system
period = 365.25 * u.day
mstar  = 1.0 * u.M_sun
distanceToHostStar(period, mstar)
```

```python
def planetBBtemp(lstar, rorbit, emissivity = 1):
    """
    Calculate the black body temperature of a planet.

    Args:
        lstar (Quantity): luminosity of the host star
        rorbit (Quantity): orbital radius
        emissivity (number): emissivity of the planet (default = 1)

    Returns:
        Black body temperature (K)
    """
    try:
        out = lstar/(emissivity * const.sigma_sb * 16 * np.pi * rorbit**2)
        out = out**(1/4)
        return out.to(u.K)
    except:
        return 0.0*u.K
```

## What planets *can* be seen?

* Desired plot: the smallest planet that can be seen for a given orbital period
    * For contrast $10^{-6}$ and $10^{-7}$
    * Want the function $R_{p}(P_{orb};C_{ir,\lambda}, L_*)$

The function for this is calculated as (see notes):
$$
R_p = \left[C_{IR,\,\lambda}\frac {L_*} {4\pi\sigma T_*^4} \frac {F_{\lambda, *}(T_*)} {F_{\lambda, P}(T_p, X)}\right]^{1/2}
$$
where
* $C_{IR, \lambda}\equiv$ contrast between star and planet in the IR over some given waveband
* $L_*\equiv$ luminosity of the host star
* $T_*\equiv$ surface temperature of the star
* $F_{\lambda, *}\equiv$ radiant intensity (radiant flux integrated over solid angle at emitting surface) of the star over some given waveband
* $T_p\equiv$ "surface" temperature of the planet
* $X\equiv$ factor accounting for exoplanet atmospheric effects. Assume this is $1$ for now.
* $F_{\lambda, p}\equiv$ radiant intensity of the planet over some given waveband

Some things that are given/assumed are
* $C_{IR, \lambda}$ determined by instrument
* $L_*$ from previous survey
* $T_*$ from spectral class
* $\lambda_\min$ determined by instrument
* $\lambda_\max$ determined by instrument
* $X = 1$, otherwise too complicated and planet-dependent. Assume that treating it as a blackbody will be a good approximation.

The radiant intensity of planets can potentially be complicated. Let's look at cases where exoplanets are self-luminous blackbodies(e.g. young gas giants) or have cooled to be in equilibrium temperature with the host star:

* If it is self-luminous then $T_p$ is given and it can be seen if $R_p$ is greater than the right side of the equation above and its orbital radius is beyond the IWA of the coronagraph. 
* Blackbody planets, on the other hand, will depend on their equilibrium temperature with the host star

$$
T_p(L_*, P_{orb})=\left\{\frac{\pi^{1/3}L_*}{4\varepsilon\sigma\left[2GM_*(L_*)P_{orb}^2\right]^{2/3}}\right\}^{1/4}
$$
where
* $\varepsilon\equiv$ planet emissivity. Let's keep this $=1$, but note that temperature increases with decreased emissivity.
* $\sigma\equiv$ the SB constant
* $M_*(L_*)\equiv$ the mass luminosity relation for main-sequence stars
    * See the `luminosityToMass` function in this notebook

We can then use Planck's law (`planckSpectralRadianceVsWavelength` in this notebook) to get the radiant intensity of the star and the planet in the instrument's waveband.

Star:

$$
F_{\lambda, *} = \pi\int_{\lambda_\min}^{\lambda_\max}B_\lambda(\lambda, T_*)d\lambda
$$

Planet:

$$
F_{\lambda, p} = \pi\int_{\lambda_\min}^{\lambda_\max}B_\lambda(\lambda, T_p)X(\lambda;\text{atmosphere stuff})d\lambda\approx \pi\int_{\lambda_\min}^{\lambda_\max}B_\lambda(\lambda, T_p)d\lambda
$$

where

$$
B_\lambda(\lambda, T) = \frac{2hc^2}{\lambda^5}\frac{1}{e^{hc/(\lambda k_B T)}-1}
$$


```python
def tempPlanetBlackbody(pOrbit, lumStar, emissivity = 1):
    """
    Treats planet as a blackbody to calculate the equilibrium temperature it will reach.

    Args:
        pOrbit  (quantity): orbital period with units of time
        lumStar (quantity): luminosity of the host star with units of power
    """

    numerator = np.pi**(1/3)*lumStar

    # get mStar
    lumStarInLsunUnits = lumStar/const.L_sun
    lumStarInLsunUnits = lumStarInLsunUnits.si.value
    mStar = luminosityToMass(lumStarInLsunUnits)*const.M_sun

    denominator = 4*emissivity*const.sigma_sb*(2*const.G*mStar*pOrbit**2)**(2/3)

    out = (numerator/denominator)**(1/4)
    
    return out.to(u.Kelvin)
```

```python
def radiusOfPlanetFromContrast(contrast, lumStar, tempStar, tempPlanet, lamMin, lamMax):
    """
    Calculates the radius of an exoplanet that can be seen for a given coronagraph contrast

    Args:
        contrast (number): contrast of the instrument
        tempStar (quantity): temperatuer of the star 
        tempPlanet (quantity): temperature of the planet
        lumStar  (quantity): luminosity of the host star
        lamMin (quantity): minimum wavelength of bandpass
        lamMax (quantity): maximum wavelength of bandpass
    """
    fluxStar = radiantFluxBandpass(tempStar, lamMin, lamMax)
    fluxPlanet = radiantFluxBandpass(tempPlanet, lamMin, lamMax)
    out = contrast * lumStar/(4*np.pi*const.sigma_sb*tempStar**4)
    out *= fluxStar/fluxPlanet
    out = np.sqrt(out)
    out = out.si
    return out
    
```

```python
lmin = 0.8 #in um
lmax = 2.5 #in um
dtel = 2.0 #in m

# Rayleigh Criterion
resAtLmin = 0.21*lmin/dtel
resAtLmax = 0.21*lmax/dtel
resAtLmid = 0.21*((lmin+lmax)/2)/dtel

# parameters
coronascopeIwaFactor = 3

res = [resAtLmin, resAtLmid, resAtLmax]
iwa = np.array([coronascopeIwaFactor * r for r in res]) # Calculate IWA for each resolution

print("Telescope Specifications:")
print(f"  Diameter: {dtel:.1f} m")
print(f"  Wavelength Range: {lmin:.1f} - {lmax:.1f} µm\n")
print("  Resolution:") # Clarify coefficient
print(f"    At {lmin:.1f} µm: {resAtLmin:.3f} arcseconds")
print(f"    At {(lmin+lmax)/2:.2f} µm: {resAtLmid:.3f} arcseconds")
print(f"    At {lmax:.1f} µm: {resAtLmax:.3f} arcseconds\n")
print("  Coronagraph Inner Working Angle (IWA):")
print(f"    At {lmin:.1f} µm: {iwa[0]:.3f} arcseconds")
print(f"    At {(lmin+lmax)/2:.2f} µm: {iwa[1]:.3f} arcseconds")
print(f"    At {lmax:.1f} µm: {iwa[2]:.3f} arcseconds\n")
print(f"Note: Coronagraph IWA is generally a multiple of the diffraction-limited resolution. Using a factor of {coronascopeIwaFactor} in this case.")
```

```python
lamMin = lmin*u.micrometer
lamMax = lmax*u.micrometer

lumStar = np.array([0.3, 1.0, 10.0])*const.L_sun
starDistArray = np.array([5, 10, 15, 20])*u.pc
contrastRange = np.logspace(-9, -3, 500)

maxAllowedPlanetRadius = 3.0

# create the figure
# plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots(len(iwa), len(lumStar), figsize = (15, 15))
for i, angle in enumerate(iwa):
    for j, lum in enumerate(lumStar):
        
        # calculate the temperature of the star
        starMass = scalarLuminosityToMass(lum/const.L_sun)
        starR = starMass**0.8 * const.R_sun
        starTemp = (lum/(4*np.pi*const.sigma_sb*starR**2))**(1/4)
        starTemp = starTemp.to(u.Kelvin)

        ax[i][j].set_xscale("log")
        ax[i][j].set_yscale("log")

        if i == (len(lumStar)-1):
            ax[i][j].set_xlabel("Contrast (unitless)")
        if j == 0:
            ax[i][j].set_ylabel("Minimum Planet Radius for Visibility ($R_{jup}$)")
        ax[i][j].set_title(f"IWA = {angle:.2f}\", L = {lum/const.L_sun} $L_\\odot$")
        for k, dist in enumerate(starDistArray):
            # calculate the minimum orbit that can be seen with the given IWA
            planetDist = (angle * (np.pi/648000) * dist).to(u.AU)
            # calculate the temperature of a planet at this orbit
            planetTemp = planetBBtemp(lum, planetDist)
            planetRbb = radiusOfPlanetFromContrast(contrastRange, lum, starTemp, planetTemp, lamMin, lamMax)/const.R_jup
            validIndices = np.where(planetRbb <= maxAllowedPlanetRadius)
            ax[i][j].plot(contrastRange[validIndices], planetRbb[validIndices], label = f"{dist}, BB")
            
        planetR500k = radiusOfPlanetFromContrast(contrastRange, lum, starTemp, 500*u.Kelvin, lamMin, lamMax)/const.R_jup
        validIndices = np.where(planetR500k <= maxAllowedPlanetRadius)
        ax[i][j].plot(contrastRange[validIndices], planetR500k[validIndices], label = f"500K")
        
        planetR1500k = radiusOfPlanetFromContrast(contrastRange, lum, starTemp, 1500*u.Kelvin, lamMin, lamMax)/const.R_jup
        validIndices = np.where(planetR1500k <= maxAllowedPlanetRadius)
        ax[i][j].plot(contrastRange[validIndices], planetR1500k[validIndices], label = f"1500K")
        ax[i][j].legend(loc = "lower right")
        ax[i][j].grid(visible=True)
```
